```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
# Create an empty DataFrame with specified columns
columns = ['R&D Spend', 'Administration', 'Marketing Spend','State','Profit']
df = pd.DataFrame(columns=columns)

print("Original DataFrame:")
print(df)

# Ask the user for the number of entries they want to provide
num_entries = int(input("How many entries do you want to add to the DataFrame? "))

for i in range(num_entries):
    entry_data = {}

    for col in columns:
        value = input(f"Enter value for column '{col}' (entry {i+1}/{num_entries}): ")
        entry_data[col] = value

    # Append the user's input to the DataFrame
    df = df.append(entry_data, ignore_index=True)

print("\nUpdated DataFrame:")
print(df)
```

```python
df
```

```python
X= df.drop('Profit',axis=1)
Y=df['Profit']
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2 , random_state=10)
```

```python
df['State'].replace(['New York', 'California','Florida'], [0, 1,2] , inplace= True)
```

```python
df
```

```python
# Train
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# predict
y_pred = model.predict(x_test)
print('The predicted sales for an advertising budget of 1000 is:', y_pred)
```

```python
# Evaluate
# calculate Mean square error
MSE = mean_squared_error(y_test,y_pred)

# calculate ROOT Mean square error
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))

# calculate  Mean absolute error
MAE=mean_absolute_error(y_test, y_pred)

# Calculate R square vale
rsq = r2_score(y_test,y_pred)

print('mean squared error :',MSE)
print('Root mean squared error :',RMSE)
print('mean absolute error :',MAE)
print('r square :',rsq)
```
